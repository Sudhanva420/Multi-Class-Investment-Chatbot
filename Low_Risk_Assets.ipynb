```python
import pandas as pd
import spacy
import re

df = pd.read_csv("safe_assets_articles_raw.csv")

#spaCy English model
nlp = spacy.load("en_core_web_sm")

#clean and normalize text
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  
    text = re.sub(r'[^\w\s]', '', text)
    text = text.lower() 
    return text.strip()

# Function to process text through spaCy- tokenize + POS tag + lemmatize
def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    pos_tags = [(token.text, token.pos_) for token in doc]
    return {
        "clean_text": " ".join(tokens),
        "tokens": tokens,
        "pos_tags": pos_tags
    }

#preprocessing
processed = df['text'].apply(lambda x: preprocess_text(clean_text(x)))

# Extract into new DataFrame columns
df['clean_text'] = processed.apply(lambda x: x['clean_text'])
df['tokens'] = processed.apply(lambda x: x['tokens'])
df['pos_tags'] = processed.apply(lambda x: x['pos_tags'])

df.to_csv("safe_assets_articles_processed.csv", index=False)
df[['asset_type', 'title', 'tokens', 'pos_tags']].head()
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

df = pd.read_csv("safe_assets_articles_processed.csv")
texts = df["clean_text"].astype(str).tolist()  # list of strings

# model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

all_embeddings = []

# Encode and pool in batches
for text in texts:
    inputs = tokenizer([text], padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
        # Mean pooling across the token embeddings
        embedding = outputs.last_hidden_state.mean(dim=1)  
        embedding_np = embedding.squeeze(0).numpy() 
        all_embeddings.append(embedding_np)

# Convert to NumPy and attach to DataFrame
embeddings_array = np.array(all_embeddings)
df["embedding"] = all_embeddings

df.to_csv("safe_assets_articles_with_embeddings.csv", index=False)

# t-SNE dimensionality reduction
tsne = TSNE(n_components=2, random_state=42, perplexity=5)
tsne_results = tsne.fit_transform(embeddings_array)

df["tsne_x"] = tsne_results[:, 0]
df["tsne_y"] = tsne_results[:, 1]

plt.figure(figsize=(10, 6))
colors = {"bonds": "blue", "gold": "gold", "silver": "gray"}
for asset in df["asset_type"].unique():
    subset = df[df["asset_type"] == asset]
    plt.scatter(subset["tsne_x"], subset["tsne_y"], 
                label=asset, c=colors.get(asset, "black"), s=60)

plt.title("t-SNE Visualization of Article Embeddings")
plt.xlabel("t-SNE X")
plt.ylabel("t-SNE Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load tokenizer and model
model_name = "microsoft/phi-1_5"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

model.eval()
```

```python
def generate_response(prompt, max_new_tokens=200):
    inputs = tokenizer(prompt, return_tensors="pt")
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=0.7,
            top_k=50
        )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
# Load same embedding model (MiniLM)
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np

tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model.eval()

def embed_query(text):
    inputs = tokenizer([text], return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1)
    return embedding.squeeze().numpy().astype("float32")
```

```python
def retrieve_relevant_articles(query, k=3):
    query_vec = embed_query(query).reshape(1, -1)
    D, I = index.search(query_vec, k)
    return df.iloc[I[0]].drop_duplicates(subset="clean_text")  # remove dupes


# Load reasoning LLM (phi-1.5), this is the generation step
llm_tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-1_5")
llm_model = AutoModelForCausalLM.from_pretrained("microsoft/phi-1_5")
llm_model.eval()

def generate_rag_answer(query, docs, max_new_tokens=200):
    context = "\n\n".join(
        [text.strip().replace("\n", " ")[:500] for text in docs["clean_text"].tolist()]
    )

    prompt = f"""You are a knowledgeable and concise financial assistant.
Use the context below to answer the user's financial question clearly and professionally.

Context:
{context}

Question: {query}
Answer:"""

    inputs = llm_tokenizer(prompt, return_tensors="pt", truncation=True, max_length=1024)
    with torch.no_grad():
        outputs = llm_model.generate(
            **inputs, max_new_tokens=max_new_tokens, early_stopping=True, do_sample=False
        )
    return llm_tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

query = "Should I invest in gold or bonds in 2024 for low risk?"
docs = retrieve_relevant_articles(query)
print("📄 Retrieved Documents:\n", docs[["title", "url"]])

answer = generate_rag_answer(query, docs)
print("💬 Final RAG Answer:\n", answer)

cur.close()
conn.close()
```

```python
eval_data = [
    {
        "query": "Should I invest in gold or bonds in 2024 for low risk?",
        "reference": "Gold is a good investment for low risk because it is a safe-haven asset that tends to hold its value during times of economic uncertainty. Bonds, on the other hand, are considered low risk as they provide a fixed income stream and are backed by the issuer's ability to repay the debt."
    },
    {
        "query": "What is the best way to diversify my investment portfolio?",
        "reference": "The best way to diversify your investment portfolio is by investing in a combination of different asset classes such as stocks, bonds, real estate, and commodities. This reduces risk and minimizes the impact of any single investment's poor performance."
    },
    {
        "query": "How can I protect my investments from inflation?",
        "reference": "To protect your investments from inflation, consider investing in assets that historically outperform inflation, such as stocks or real estate. Another option is to invest in Treasury Inflation-Protected Securities (TIPS), which are specifically designed to hedge against inflation."
    }
]
```

```python
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer

bleu_scores = []
rouge_l_scores = []

for item in eval_data:
    query = item["query"]
    reference = item["reference"]

    # RAG step
    docs = retrieve_relevant_articles(query)
    prediction = generate_rag_answer(query, docs)

    # BLEU
    bleu = sentence_bleu([reference.split()], prediction.split())
    bleu_scores.append(bleu)

    # ROUGE-L
    scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    rouge = scorer.score(reference, prediction)["rougeL"].fmeasure
    rouge_l_scores.append(rouge)

    print(f"\n🟢 Query: {query}")
    print(f"📌 Prediction: {prediction}")
    print(f"✅ Reference: {reference}")
    print(f"🔵 BLEU: {bleu:.3f} | 🔴 ROUGE-L: {rouge:.3f}")
```

```python
print("\n📈 AVERAGE BLEU:", round(np.mean(bleu_scores), 3))
print("📈 AVERAGE ROUGE-L:", round(np.mean(rouge_l_scores), 3))
```
